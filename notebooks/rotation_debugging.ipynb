```python
import cv2
import numpy as np
import glob
import os

img = cv2.imread('/Users/nathaning/_original_data/pca/seg_0.1/feat/test1sss.jpg')

%matplotlib inline
from matplotlib import pyplot as plt
```

```python
rotations = [img]
de_rotations = []

for k in range(1,4):
    rotations.append(np.rot90(img, k))
    de_rotations.append(np.rot90(rotations[k-1], 4-k))
    
_, axes_1 = plt.subplots(1,4)
for k, ax in enumerate(axes_1):
    ax.imshow(rotations[k])
    
_, axes_2 = plt.subplots(1,3)
for k, ax in enumerate(axes_2):
    ax.imshow(de_rotations[k])
```
